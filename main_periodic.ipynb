```python
import gmsh
import numpy as np



# 
# 1. Create a new model
gmsh.initialize()

gmsh.model.add("composite")

# 2. Create circle using occ

# Create a circle with center (x0, y0, z0) and radius r
ry = 0.5
r  = 5


# ==================================================


p_1 = gmsh.model.occ.addPoint(r, 0, 0)
p_2 = gmsh.model.occ.addPoint(0, r, 0)
p_0 = gmsh.model.occ.addPoint(0, 0, 0)

gmsh.model.occ.synchronize()

arc1 = gmsh.model.occ.addCircleArc(p_1, p_0, p_2)
wr_arc_1 = gmsh.model.occ.addWire([arc1])

# synchronize
gmsh.model.occ.synchronize()
# add pipe
rc_1 = gmsh.model.occ.addDisk(0,r,0, 
                                ry, 
                                ry,
                                zAxis=[1,0,0],
                                xAxis=[0,0,1]) 
pipe_1 = gmsh.model.occ.addPipe([(2,rc_1)], wr_arc_1)
gmsh.model.occ.synchronize()

gmsh.fltk.run()
# ==================================================



p_1 = gmsh.model.occ.addPoint(3*r, 0, 0)
p_2 = gmsh.model.occ.addPoint(4*r, r, 0)
p_0 = gmsh.model.occ.addPoint(4*r, 0, 0)

gmsh.model.occ.synchronize()

arc2 = gmsh.model.occ.addCircleArc(p_1, p_0, p_2)
wr_arc_2 = gmsh.model.occ.addWire([arc2])

rc_2 = gmsh.model.occ.addDisk(4*r,r,0,
                                ry, 
                                ry,
                                zAxis=[1,0,0] ,
                                xAxis=[0,0,1])

# synchronize
gmsh.model.occ.synchronize()

pipe_2 = gmsh.model.occ.addPipe([(2,rc_2)], wr_arc_2)

# ==================================================

p_1 = gmsh.model.occ.addPoint(r, 2*r - 1e-10, 0)
p_2 = gmsh.model.occ.addPoint(3*r, 2*r, 0)
p_0 = gmsh.model.occ.addPoint(2*r, 2*r, 0)

gmsh.model.occ.synchronize()

arc3 = gmsh.model.occ.addCircleArc(p_1, p_0, p_2)
wr_arc_3 = gmsh.model.occ.addWire([arc3]) 
# 
# box
gmsh.model.occ.synchronize()

rc_3 = gmsh.model.occ.addDisk(r,2*r,0,
                                ry, 
                                ry,
                                zAxis=[0,-1,0],
                                xAxis=[0,0,1])

pipe_3 = gmsh.model.occ.addPipe([(2,rc_3)], wr_arc_3)
gmsh.model.occ.synchronize()

gmsh.write("yarns.brep")

gmsh.finalize()

# ==================================================



gmsh.initialize()
gmsh.open("yarns.brep")
gmsh.model.occ.synchronize()


volume_yarn = gmsh.model.getEntities(3)
zL = 3*ry 

box = gmsh.model.occ.addBox(0   ,   0   , -zL/2, 
                            4*r ,   2*r , zL)
# cut # synchronize
gmsh.model.occ.synchronize()

cut = gmsh.model.occ.cut(volume_yarn, [(3, box)], removeObject=False, removeTool=True)


gmsh.option.setNumber("Mesh.CharacteristicLengthMin", ry*0.1)
gmsh.option.setNumber("Mesh.CharacteristicLengthMax", 2*ry)
gmsh.option.setNumber("Mesh.MeshSizeFromCurvature", 10)
gmsh.option.setNumber("Mesh.AngleSmoothNormals", 10)
gmsh.option.setNumber("Mesh.Smoothing", 10)	
gmsh.option.setNumber("Mesh.Algorithm", 2)
# synchronize

gmsh.model.mesh.generate(3)

gmsh.model.mesh.setOrder(2)

gmsh.write("periodic.inp")

```

```python
gmsh.fltk.run()
```
