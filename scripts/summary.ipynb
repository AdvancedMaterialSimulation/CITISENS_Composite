```python
import glob,os
import pandas as pd
from matplotlib import pyplot as plt
```

```python
files = glob.glob('Homo*/*.csv',recursive=True)
```

```python
files
```

```python
dfs = [ pd.read_csv(f) for f in files]
files = [ f.replace("Homogenization_","").split(os.sep)[0] for f in files ]
```

```python
ms = [ 11,11,10,9]
for idf,ifile,ims in zip(dfs,files,ms):
    plt.plot(100*idf["vfrac"],1e-3*idf["E"],"o-",label="design " + ifile)
plt.xlabel("Volume Fraction [%]")
plt.ylabel("E [GPa]")
plt.legend()
plt.grid()
```
