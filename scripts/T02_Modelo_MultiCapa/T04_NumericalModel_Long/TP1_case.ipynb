```python
B = 25
x0 = 0
y0 = B/2
eps = 0.1
data = ifrd["data"]
data = data[data["x"] > x0 - eps]
data = data[data["x"] < x0 + eps]
data = data[data["y"] > y0 - eps]
data = data[data["y"] < y0 + eps]
```

```python
# sort z 
data = data.sort_values(by="z")
```

```python
from matplotlib import pyplot as plt
```

```python
plt.plot(data["z"], data["P1"], label="P1", marker="o")

t_l = 0.427120
t_n = 3.073400

y1 = -t_n/2
plt.axvline(y1, color="red", linestyle="--", label="t_n/2")
y2 = -t_n/2 - t_l
plt.axvline(y2, color="red", linestyle="--", label="t_n/2 + t_l")
y3 = -t_n/2 - 2*t_l
plt.axvline(y3, color="red", linestyle="--", label="t_n/2 + 2*t_l")
```
