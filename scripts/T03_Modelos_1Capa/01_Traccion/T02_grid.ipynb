```python
#import plot 
from matplotlib import pyplot as plt
import numpy as np
```

```python
d = 16
r = 2
rho = 4

lx = d + 2*r
np_x = lx*rho 
x_span = np.linspace(-(d+2*r)/2,(d+2*r)/2,np_x)

ly = 2*r
np_y = ly*rho
y_span = np.linspace(-r,r,np_y)

X,Y = np.meshgrid(x_span,y_span)
X = X.flatten()
Y = Y.flatten()

plt.plot(X,Y,".")

idx = ((X-d/2)**2 + Y**2 > r**2) 
idx_2 = X > d/2 

idx = idx * idx_2
X = X[~idx]
Y = Y[~idx]

idx = ((X+d/2)**2 + Y**2 > r**2) 
idx_2 = X < -d/2 

idx = idx * idx_2
X = X[~idx]
Y = Y[~idx]

plt.plot(X,Y,".")

plt.gca().set_aspect(1)
```
