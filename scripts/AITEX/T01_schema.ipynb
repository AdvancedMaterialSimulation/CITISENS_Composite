```python
nucleo = 0.1111 # in
composite = 0.191
# to mm
nucleo = nucleo * 25.4
composite = composite * 25.4

print("Nucleo: ", nucleo)
print("Composite: ", composite)

total_layer = (composite - nucleo)/2

one_layer = total_layer/2
print("Total layer: ", total_layer)
print("One layer: ", one_layer)
```

```python
from matplotlib import pyplot as plt

```

```python
# rectangle
def rectangle(x0,x1,y0,y1,label=None,lw=None):
    plt.plot([x0,x1,x1,x0,x0],[y0,y0,y1,y1,y0],label=label,lw=lw)

plt.figure()
x0 = 0
xL = 2
rectangle(x0,xL, -composite/2 ,composite/2 ,label="Composite",lw=4)
rectangle(x0+0.04,xL-0.04, -nucleo/2    , nucleo/2   ,label="Nucleo",lw=4)

rectangle(x0+0.04,xL-0.04, 
          nucleo/2 + 0.1, 
          nucleo/2+total_layer - 0.1,
          label="Layers",lw=4)

# text size of layers
plt.text(x0+0.1, nucleo/2+0.25, "Layers Size: {:.2f} mm".format(total_layer),fontsize=12) 

# nucleo
plt.text(x0+0.1, nucleo/2-1., "Nucleo Size: {:.2f} mm".format(nucleo),fontsize=12)
# x = 0 line
plt.plot([x0,xL],[0,0])

# total 
plt.text(x0+0.1, -2, "Total Size: {:.2f} mm".format(composite),fontsize=12)
plt.legend()
```
