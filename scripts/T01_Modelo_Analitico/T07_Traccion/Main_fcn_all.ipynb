#### 1. Thickness Model
$$ \Tau (t_n,t_l,n_i) = t_n + 2t_l n_i $$
#### 2. Tensile Model

$$ \mathcal{E}_t(E_n,\vec{E_l},t_n,t_l, \mathcal{C}) = 
    \frac{E_n t_n + 2t_l \  ( \vec{\delta} \cdot \vec{E_l})}{t_t} $$
where: 
- $t_t = \Tau(t_n,t_l,n_i)$
- $n_i$ is the number of layers.
- $\delta = \mathcal{F}(\mathcal{C})$ is the function of the # of layers of each type.

#### 3. Bending Model
$$ \mathcal{E}_b(E_n,\vec{E_l},t_n,t_l) = 
\frac{E_n I_n + 2 \  (\vec{I_l} \cdot \vec{E_l})}{I_t} $$

```python

import scipy.optimize as opt
from optim.objetive import objetive
from validation.experimental import manual_data


factor={"Et":1,"Eb":0,
        "St":1,"Sb":0}

t_nucleo_exp = manual_data()["t_nucleo [mm]"]

tn = 2.934
tl = 0.475

# FEM Prediction
# E_X   = 1.2183454466962444
# E_SX  = 1.1286480552634521
# E_Y   = 5.987016815409205
# E_SY  = 1.1340256789266567

# optimazation results
E_X = 1.58
E_SX = 0.92 
E_Y = 6.39 
E_SY = 1.06 

E_n = 0.80
# Mean yarn
ratios  = {'SX': np.float64(3.7370129639776195),
 'SY': np.float64(4.046305929413829),
 'X': np.float64(1.2844055304827087),
 'Y': np.float64(2.5841400155862795)}
# Max model
ratios = {'X': np.float64(1.355466057012968),
 'Y': np.float64(2.585327638119705),
 'SX': np.float64(19.070390048976318),
 'SY': np.float64(15.105436178519154)}

def f(x):
    El = np.array([E_X, E_SX, E_Y, E_SY])
    # 
    # X SX Y SY
    P1_rotura = x[0]

    Sl = P1_rotura*np.array([1/ratios["X"],
                             1/ratios["SX"],
                             1/ratios["Y"],
                             1/ratios["SY"]])


    return objetive(tn, tl, En, El,Sl,df,ni,factor=factor)


Sl_0 = [1]

x0 = [*Sl_0]
x0 = np.array(x0)


eps = 1e-3

bounds = [ 
           (eps,None),  # P1_rotura
           ]  

res = opt.basinhopping(f, x0, niter=20, 
                       minimizer_kwargs={"bounds":bounds, 
                                         "method":"Nelder-Mead"})
```

```python
xopt = res.x

En_opt = E_n
El_opt = np.array([E_X, E_SX, E_Y, E_SY])
P1_rotura = xopt[0]

Sl_opt = P1_rotura*np.array([1/ratios["X"],
                                1/ratios["SX"],
                                1/ratios["Y"],
                                1/ratios["SY"]])

df_pred,error_df = objetive(tn, tl, 
                            En_opt, El_opt, Sl_opt,
                            df,ni,give_pred=True,factor=factor)
```
