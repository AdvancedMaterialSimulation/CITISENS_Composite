```python

# wrrite 
import pandas as pd

df_01 = pd.read_csv('data.txt', sep='\t', header=None, names=['E tracción', 'E flexion','sigma tracción', 'sigma flexion'])
df_02 = pd.read_csv('data_02.txt', sep='\t', header=None, names=['E tracción', 'E flexion','sigma tracción', 'sigma flexion'])
```

```python
from matplotlib import pyplot as plt    
```

```python
# crear una figura de barras donde se pueda ver la diferencian entre tracción y flexión para las dos metodologías 
# las filas son nombradas TFP1, TFP2, TFP3, TFP4, TFP5, TFP6
# E esta medido en GPa y sigma en MPa

df_01.index = ['TFP1', 'TFP2', 'TFP3', 'TFP4', 'TFP5', 'TFP6']
df_02.index = ['TFP1', 'TFP2', 'TFP3', 'TFP4', 'TFP5', 'TFP6']


E_traccion_mtd_01 = df_01['E tracción']
E_flexion_mtd_01 = df_01['E flexion']
E_traccion_mtd_02 = df_02['E tracción']
E_flexion_mtd_02 = df_02['E flexion']

# bar plot
fig, ax = plt.subplots(figsize=(8, 5))
width = 0.15  # the width of the bars
x = range(len(df_01.index))
# Create bars for the first methodology
bars1 = ax.bar(x , E_traccion_mtd_01, width, label=r'$E_{tracción}$  Metod. 01')
bars3 = ax.bar([p + width  for p in x], E_traccion_mtd_02, width, label=r'$E_{tracción}$  Metod. 02')

bars2 = ax.bar([p + width *2 for p in x], E_flexion_mtd_01, width, label=r'$E_{flexión}$    Metod. 01')
# Create bars for the second methodology
bars4 = ax.bar([p + width * 3 for p in x], E_flexion_mtd_02, width, label=r'$E_{flexión}$    Metod. 02')

# ticks center 
ax.set_xticks([p + 1.5 * width for p in x])
ax.set_xticklabels(df_01.index)
plt.legend()
plt.grid()
plt.ylabel('E (GPa)')
```

```python
S_traccion_mtd_01 = df_01['sigma tracción']
S_flexion_mtd_01 = df_01['sigma flexion']
S_traccion_mtd_02 = df_02['sigma tracción']
S_flexion_mtd_02 = df_02['sigma flexion']
# bar plot
fig, ax = plt.subplots(figsize=(8, 5))
width = 0.15  # the width of the bars
x = range(len(df_01.index))
# Create bars for the first methodology
bars1 = ax.bar(x , S_traccion_mtd_01, width, label=r'$σ_{tracción}$  Metod. 01')
bars3 = ax.bar([p + width  for p in x], S_traccion_mtd_02, width, label=r'$σ_{tracción}$  Metod. 02')
bars2 = ax.bar([p + width *2 for p in x], S_flexion_mtd_01, width, label=r'$σ_{flexión}$    Metod. 01')

# Create bars for the second methodology
bars4 = ax.bar([p + width * 3 for p in x], S_flexion_mtd_02, width, label=r'$σ_{flexión}$    Metod. 02')
# ticks center
ax.set_xticks([p + 1.5 * width for p in x])
ax.set_xticklabels(df_01.index)
plt.legend()
plt.grid()
plt.ylabel('σ (MPa)')
```
