```python
# load fourier_cylinder_occ_pipe.stl

import gmsh

gmsh.initialize()
gmsh.model.add("fourier_cylinder_occ_pipe")


h = 3

gmsh.open("fourier_cylinder_occ_pipe.brep")

volume_yarn = gmsh.model.getEntities(3)[-1]
# tranlase volume_yarn in y direction
# add box 
rep = 2
box = gmsh.model.occ.addBox(1   , -h/2 , -0.125, 
                            4*rep, h,  0.25)

# gmsh.model.occ.addBox(0.5, -2, -1, 
#                       5, 4,  2)

gmsh.model.occ.synchronize()
# cut box - cylinder

gmsh.model.occ.cut([(3,box)], [volume_yarn], 
                    removeObject=True,
                    removeTool=False)

gmsh.model.occ.synchronize()



# repetitions 
boxes = [box]
for i in range(1, 4):
    yarn_copy = gmsh.model.occ.copy([volume_yarn])
    gmsh.model.occ.translate(yarn_copy
                             , 0, h*i, 0)
    box = gmsh.model.occ.addBox(1   , -h/2 + h*i , -0.125,
                                4*rep, h       ,   0.25)
    
    gmsh.model.occ.cut([(3,box)], [volume_yarn], 
                        removeObject=True,
                        removeTool=False)
    gmsh.model.occ.synchronize()
    boxes.append(box)

# fuse boxes
# gmsh.model.occ.fuse( [(3,boxes[0])], [(3,boxes[1])])
# gmsh.model.occ.fuse( [(3,boxes[2])], [(3,boxes[3
#                                                ])])

# iterative fuse- fuse the children with the other box of the list
if len(boxes) == 2:
    gmsh.model.occ.fuse( [(3,boxes[0])], [(3,boxes[1])])
else:
    box_union = gmsh.model.occ.fuse( [(3,boxes[0])], [(3,boxes[1])])
    box_union = box_union[0]
    for i in range(2, len(boxes)):
        box_union = gmsh.model.occ.fuse( box_union, [(3,boxes[i])])
        box_union = box_union[0]

# remove 2d entities
for i in gmsh.model.getEntities(2):
    gmsh.model.occ.remove([i])

gmsh.fltk.run()


gmsh.option.setNumber("Mesh.CharacteristicLengthMin", 0.05)
gmsh.option.setNumber("Mesh.CharacteristicLengthMax", 0.1)
gmsh.option.setNumber("Mesh.AngleToleranceFacetOverlap", 0.1)

gmsh.model.occ.synchronize()


#
gmsh.fltk.run()
# meshing 0.1
gmsh.model.mesh.generate(3)
gmsh.model.mesh.setOrder(2)

gmsh.fltk.run()

gmsh.write("fourier_cylinder_occ_pipe.inp")
```
